```python
import pandas as pd
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Load your map data
file_path = '/content/drive/My Drive/Intro to Machine Learning 4105/map_data.csv'
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
import folium

# Create a base map
m = folium.Map(location=[0, 0], zoom_start=2)

# Loop through data points and add circles
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row.get('value', 5),  # size based on value (or use a fixed number)
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(m)

# Show map
m
```

```python
m.save("map.html")
```
