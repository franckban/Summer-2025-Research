```python
# 2. Mount Google Drive and load Excel file
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

# Load Excel file (update sheet_name if needed)
file_path = '/content/drive/My Drive/Intro to Machine Learning 4105/WW Marine Datashare.xlsx'
df = pd.read_excel(file_path, sheet_name=0)  # Change sheet_name if needed

# Preview the dataset
df.head()
```

```python
import folium

# Base map centered at the equator
m = folium.Map(location=[0, 0], zoom_start=2)

# Add bubbles to map
for _, row in df.iterrows():
    try:
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],  # use actual column names
            radius=row.get('Count', 5) * 0.5,               # adjust size scaling if needed
            color='blue',
            fill=True,
            fill_opacity=0.6,
            popup=f"{row.get('Name', 'Location')} - {row.get('Count', '')}"
        ).add_to(m)
    except:
        continue  # skip bad rows

m
```
