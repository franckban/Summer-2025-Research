```python
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load CSV
file_path = '/content/drive/My Drive/Intro to Machine Learning 4105/map_data.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df['longitude'], df['latitude'], c='blue', s=10)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Map Data Points")
plt.grid(True)
plt.show()
```

```python
!pip install cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

# Create map with coastlines
fig = plt.figure(figsize=(12,6))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.scatter(df['longitude'], df['latitude'], color='red', s=10, transform=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.set_title("Geographic Data Points")
plt.show()
```
