```python
# Install required packages
!pip install folium openpyxl

# Import libraries
import pandas as pd
import folium
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load Excel data
file_path = '/content/drive/My Drive/Intro to Machine Learning 4105/WW Freshwater Datashare.xlsx'
df = pd.read_excel(file_path, sheet_name=0)

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Rename lat/lon if needed
df = df.rename(columns={
    'sample lat/long - latitude': 'latitude',
    'sample lat/long - longitude': 'longitude'
})

# Drop rows with missing or non-numeric lat/lon
df = df.dropna(subset=['latitude', 'longitude'])
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
df = df.dropna(subset=['latitude', 'longitude'])

# Create map
m = folium.Map(location=[0, 0], zoom_start=2)

# Add data points to map
for _, row in df.iterrows():
    try:
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=5,
            color='blue',
            fill=True,
            fill_opacity=0.6
        ).add_to(m)
    except:
        continue

# Show map
m
```
